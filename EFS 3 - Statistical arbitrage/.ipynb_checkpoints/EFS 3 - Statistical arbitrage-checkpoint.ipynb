## <ul><font size = "4"><b><u>Imports</u></b></font>

<ul>The code reads the CSV file using the "<code>read_csv</code>" function.

<u>Instruction</u>: Replace the "<code>...</code>" with "<code>GLD.csv</code>", "<code>GDX.csv</code>" and "<code>USO.csv</code>" in the below cell.

```python
import urllib, numpy, pandas, matplotlib.pyplot
%matplotlib inline

quotes = {"x": "GLD", "y": "GDX", "z": "USO"}
repo = "https://raw.githubusercontent.com/gsolaril/Trading_EPAT_Python/" \
       + "master/EFS%203%20-%20Statistical%20arbitrage/"
csvs = []

for var in quotes.keys():
    filename = quotes[var] + ".csv"
    urllib.request.urlretrieve(repo + filename, filename)
    csvs.append(pandas.read_csv(filename, index_col = 0)['Adj Close'])

DF = pandas.concat(csvs, axis = 1).round(2)
DF.columns = quotes.values()
DF.index = pandas.to_datetime(DF.index)

matplotlib.pyplot.style.use("dark_background")
(DF.pct_change() + 1).cumprod().plot(figsize = (10, 5))
matplotlib.pyplot.xlabel(xlabel = "Date", fontweight = "bold")
matplotlib.pyplot.ylabel(ylabel = "Percentage Change", fontweight = "bold")
matplotlib.pyplot.show()
```

## <ul><font size = "4"><b><u>Hedge ratio</u></b></font>

<ul>Calculations for Hedge Ratio on each explanatory variable/quote below, based on the initial 90 days ("<code>hdays</code>").<br>Explained variable/quote ("<code>y</code>") will be "GLD" (1st column of DataFrame, "<code>[0]</code>")

```python
import statsmodels.api
hdays = 90
y = DF.columns[0] ; x = list(DF.columns) ; x.remove(y)
model = statsmodels.api.OLS(DF[y].iloc[:hdays], DF[x].iloc[:hdays])
model = model.fit()
print('Hedge ratios...')
for quote in model.params.index:
    print(" -> %s vs. %s: %.4f" % (quote, y, model.params[quote]))
```

## <ul><font size = "4"><b><u>Spread</u></b></font>

<ul>Spread is formed as "$\Delta S = C_{y} - \Sigma_{x \neq y} \; h_{x} C_{x} $" where "$h_{x}$" is the hedge ratio for quote "$x$" and "$C_{x}$ is its price.<br>"$y$" is a particular quote in which we want to cointegrate the others.<br>In this particular case, we use: "$\Delta S = C_{[GLD]} - h_{[GDX]} C_{[GDX]} - h_{[USO]} C_{[USO]}$".<br>Series "<code>model.params[0]</code>" stores "$h_{[GDX]}$" and "<code>model.params[1]</code>" stores "$h_{[USO]}$".

```python
DF['Spread'] = DF[y] - numpy.dot(DF[x], model.params)
DF.head()
```

```python
sstr = "y[%s]" % y
for quote in model.params.index:
    hr = model.params[quote]
    sstr += " - %.4f x[%s]" % (hr, quote)
# Plot the spread
print("The spread is calculated as  --------->  %s" % sstr)
matplotlib.pyplot.style.use("dark_background")
DF["Spread"].plot(figsize = (10, 5))
matplotlib.pyplot.ylabel("Spread")
matplotlib.pyplot.show()
```

## <ul><font size = "4"><b><u>ADF Test</u></b></font>

<ul>We determine the spread is cointegrated using "<code>adfuller</code>" method, with "<code>maxlag = 1</code>".

```python
# Compute ADF test statistics
adf = statsmodels.tsa.stattools.adfuller(DF["Spread"], maxlag = 1)
print("T-stat for 1st root:", adf[0])
print("Null hypotesis prob:", adf[4])
```

<ul>If "T-stat" value is less than the critical value then the spread is cointegrated.<br>Cell below should give "<code>True</code>" if the spread is cointegrated and "<code>False</code>" if it's not.<br>(Assume 90% confidence level)

```python
is_triplet_cointegrated = (adf[4][list(adf[4].keys())[-1]] <= -1)
print("Is triplet cointegrated? %s" % is_triplet_cointegrated)
```

## <ul><font size = "4"><b><u>Mean reversion</u></b></font>

<ul>Input parameters:
<ul> <li>"$K$" ("<code>std_dev</code>"): Number of standard deviations that define Bollinger bandwidth.
</li><li>"$P$" ("<code>lookback</code>"): Amount of candles being looked back for their calculation. Also called "period".

```python
def stat_arb(df, lookback, std_dev):

    # Calculation of moving average and moving standard dev.
    df['moving_average'] = df["Spread"].rolling(lookback).mean()
    df['moving_std_dev'] = df["Spread"].rolling(lookback).std()
    # Calculation of Bollinger upper and lower bands
    df['upper_band'] = df['moving_average'] + std_dev*df.moving_std_dev
    df['lower_band'] = df['moving_average'] - std_dev*df.moving_std_dev
    # Condition for buying: price below lower band.
    df['long_entry'] = df["Spread"] < df['lower_band']
    # Condition for selling: increasing price reverted to mean.
    df['long_exit'] = df["Spread"] >= df['moving_average']
    # Seek for all long positions, between entry and exit.
    df['positions_long'] = numpy.nan
    df.loc[df['long_entry'], 'positions_long'] = 1
    df.loc[df['long_exit'],  'positions_long'] = 0
    df['positions_long'] = df['positions_long'].fillna(method = 'ffill')
    # Condition for selling: price above upper band.
    df['short_entry'] = df["Spread"] > df['upper_band']
    # Condition for buying: decreasing price reverted to mean.
    df['short_exit'] = df["Spread"] <= df['moving_average']
    # Seek for all short positions, between entry and exit.
    df['positions_short'] = numpy.nan
    df.loc[df['short_entry'], 'positions_short'] = -1
    df.loc[df['short_exit'],  'positions_short'] = 0
    df['positions_short'] = df['positions_short'].fillna(method = 'ffill')
    # Join all position flags.
    df['positions'] = df['positions_long'] + df['positions_short']
    # Get spread increment.
    df['spread_difference'] = df["Spread"] - df["Spread"].shift(1)
    # Get profit and loss.
    df['PNL'] = df['positions'].shift(1) * df['spread_difference']
    df['Cumulative profit'] = df['PNL'].cumsum()

    return df
```

<ul>Call the function "<code>stat_arb</code>" with "<code>df = DF</code>", "<code>lookback = 15</code>" and "<code>std_dev = 1</code>".

```python
DF = stat_arb(df = DF, lookback = 15, std_dev = 1)
```
