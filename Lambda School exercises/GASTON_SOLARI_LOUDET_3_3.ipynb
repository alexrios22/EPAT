### Examine total bill and tip to identify any correlation. Calculate the correlation coefficient, AND plot both features as a scatterplot.

```
def corr_scatter(X = "total_bill", Y = "tip"):
  Data.plot.scatter(x = X, y = Y)
  plt.grid(True)
  Labels = {"total_bill" : "Bill amount",
                  "size" : "Table size",
                   "tip" : "Tip amount"}
  plt.xlabel(Labels[X])
  plt.ylabel(Labels[Y])
  plt.title("%s-%s scatterplot and regression line" \
            % (Labels[X], Labels[Y]))
  x = np.array([0, Data[X].max()])
  a0, a1 = polyfit(Data[X], Data[Y], 1)  # Calculate regression line coeffs.
  plt.plot(x, a0 + a1*x, color = 'r', linewidth = 2)  # Plot regression line.
  plt.show()
  print("The correlation coefficient between '%s' and '%s' is: %0.4f" \
        % (Labels[X], Labels[Y], Data.corr()[X][Y]))
```

```
corr_scatter(X = "total_bill", Y = "tip")

print("Moderate positive correlation.\nVariability of tip sizes growing " \
      "with larger bills.")
```

### Examine size and total bill to identify any correlation. Calculate the correlation coefficient, AND plot both features as a scatterplot.
