### Import Numpy as Np

```
import numpy as np
```

### Generate random numbers
Create a (5,3) Numpy array of random integer values between 0 and 100.

```
R = 100.0*np.random.random((5, 3))
print(R)
```

### Array practice
Run each of these in their own code cell and comment it accordingly:

*  Return the first row
*  Return the last column
*  Return the third column values from the 4th and 5th rows
*  Multiply every value in the array by 2
*  Divide every value by 3
*  Increase the values in the first row by 12
* Calculate the mean of the first column
* Calculate the median of the array _after_ removing the 2 smallest values in the array
* Calculate the standard deviation of the first 3 rows
* Return values in the second column greater than 25
* Return values in the array under 40

```
# Return the first row:
R_01 = R[0, :]
print("First row: \n" + str(R_01))
```

```
# Return the last column:
R_02 = R[:, -1]
print("Last column: \n" + str(R_02))
```

```
# Return the third column values from the 4th and 5th rows:
R_03 = R[3:5, 2]
print("3rd column from row 4th & 5th: \n" + str(R_03))
```

```
# Multiply every value in the array by 2:
R_04 = R*2.0
print("Multiply values in array by 2: \n" + str(R_04))
```

```
# Divide every value by 3:
R_05 = R/3.0
print("Divide values in array by 3: \n" + str(R_05))
```

```
# Increase the values in the first row by 12:
R_06 = R.copy() # Do not affect initial random array.
R_06[0] += 12
print("Add 12 to first row values: \n" + str(R_06))
```

```
# Calculate the mean of the 1st column:
R_07 = np.mean(R[:, 0])
print("Calculate mean of 1st column: \n" + str(R_07))
```

```
# Median of the array after removing the 2 smallest values:
R_08a = np.reshape(R.copy(), -1)  # Reshape R into 1D array.
R_08b = sorted(R_08a)[2:]  # Sort from small to large; erase 2 leftmost values.
R_08  = np.median(R_08b)    # Calculate median
print("Array median without 2 smallest values: \n" + str(R_08))
```

```
# Calculate the standard deviation of the first 3 rows:
R_09 = np.std(R[:3, :])
print("Standard deviation of first 3 rows: \n" + str(R_09))
```

```
# Return values in the second column greater than 25:
R_10a = R[:, 1]
R_10  = R_10a[R_10a > 25]
print("Values in 2nd column > 25: \n" + str(R_10))
```

```
# Return values in the array under 40:
R_11 = R[R < 40]
print("values in whole array < 40: \n" + str(R_11))
```
