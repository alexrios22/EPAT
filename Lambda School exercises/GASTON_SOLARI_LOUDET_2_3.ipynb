### Import pandas

```
import pandas as pd
```

### Read in your data
Your dataset can be found [here.](https://raw.githubusercontent.com/axrd/datasets/master/tipdata.csv)
Import it and return the first 5 rows:

```
Data = pd.read_csv("https://raw.githubusercontent.com/axrd/datasets/master/tipdata.csv")
Data.head()
```

### What is the shape of the DataFrame?

```
Ds = Data.shape
print("%d rows, %d columns" % Ds)
```

### Get specific values

```
# What is the gender of the server in the 41st entry?
print("The 41st entry's gender is %s." % Data["sex"].iloc[40])
```

```
# Was the 22nd entry seated in a smoking section?
x = "not " if (Data["smoker"].iloc[21] == "No") else ""
print("The 22nd was %sseating in smokers' section." % x)
```

### Get a range of values

```
# Return the last 20 entries.
Data.tail(20)
```

```
# Return entries 25 through 50.
Data.head(51).tail(26)
```

### Filtering and subsetting:

```
# Return all the entries where the size is greater than 2.
Data.where(Data["size"] > 2).dropna()
```

```
# Create a DataFrame consisting only of data for Fridays. 
# Hint: You should have 19 values.
DF = Data.where(Data["day"] == "Fri").dropna()
print("Fridays' dataframe has %d entries.\n" % DF["day"].size)
DF.head(200000) # For any n > 19, dataframe size is the same.
```
