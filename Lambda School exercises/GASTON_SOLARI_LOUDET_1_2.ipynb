### 1.3 We need to calculate Joey's body mass index. Here's the formula for BMI:  

>$\frac{Weight(kg)}{Height(m)^2} = BMI$  

Joey is 1.82 meters tall and weighs 78 kg. Create a variable for his height, weight, and BMI by combining the first two. 


```
H, W = 1.82, 78
BMI = W/(H**2)
print("Joey's body mass index is %0.2f kg/m^2." % BMI)
```

## Part 2: Data Types

### 2.1 What is the data type of the following variables?

```
a = -91
b = 12.0222
c = True
d = "Loosey goosey"

Names = ["a", "b", "c", "d"]
Values = dict(zip(Names, [a, b, c, d]))
for Var in Names:
  Type = type(Values[Var])
  print("Variable %s's data type is: %s" % (Var, Type))
```

### 2.2 How do strings behave differently than integers or floats?

*Write your answer below this line*: 

---
**Short answer: Integers represent whole numbers, while floats represent rational numbers (can store decimal places).**

The integer is digitally stored in memory such that the leftmost binary digit represents the sign, and the rest of the digits represent the absolute value. Can be stored in different sizes (8, 16, 32 or 64 bits).
Floats use floating point precision. Leftmost binary digit is the sign. The rest of them represent an exponent and a base, very much akin to the scientific notation. Hence, larger numbers have less precision. Can also be stored in different sizes (mostly 32 bits onwards).
