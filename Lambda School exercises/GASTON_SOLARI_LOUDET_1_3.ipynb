### 1.2 Combine the three lists into a new list where each element is a person's name, weight, and height. Call it "person_data".

```
person_data = []
for n, name in enumerate(names):
  element = [name, weights[n], heights[n]]
  person_data.append(element)
print(person_data)
```

### 2.2 Use a for loop to calculate the BMI of each person.
>$\frac{Weight(kg)}{Height(m)^2} = BMI$  

```
fun = lambda weight, height : weight/(height**2)
for n in range(len(person_data)):
  weight = person_data[n][1]
  height = person_data[n][2]/100.0 # Convert to metres.
  BMI.append(fun(weight, height))
print(BMI)
```

### 2.2 BONUS: Do it using list comprehension.

```
BMI = [fun(person[1], person[2]/100.0) for person in person_data]
print(BMI)
```
