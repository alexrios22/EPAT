### 1.4 Create a function that returns the sum of two numbers.

```
def add_two_numbers(x1, x2):
  """Arithmetically adds up both arguments."""
  return(x1 + x2)

print(add_two_numbers(2, 3))
```

### 1.5 Create a function that, given a list of numbers, returns their average.

```
def average(list):
  return(sum(list)/len(list))

print(average([2.5, 3.0, 3.5]))
```

### 1.6 Create a function that, given a string, returns that string capitalized and reversed. 

```
def caprev(string):
  caprev = ""
  for char in range(len(string)):
    caprev += string[-(1 + char)].upper()
  return(caprev) 

print(caprev("!elbativeni ma ...I"))
```

### 1.7 Create a function that, given a list of strings, concatenates them together with proper spacing.

```
def concat(strings):
  joined = ""
  for string in strings:
    if string == strings[-1]:
      joined += string + "." # End of sentence
    else:
      joined += string + " " # Spacing
  return(joined)

print(concat(["And", "I...", "am", "Iron", "Man"]))
```
